```python
from scipy.stats import norm
```

```python
val,m,s = 68,65.5,2.5 # val=x, mean=m, std dev=s 
```

```python
print(norm.cdf(val,m,s))
```

```python
# cdf(x>val)
print(1-norm.cdf(val,m,s))
```

```python
# cdf(val1<x<val2)
print(norm.cdf(val,m,s)-(norm.cdf(63,m,s)))
```

```python
# P(x>700 | m=494 amd s=100)=?
print(1-norm.cdf(700,494,100))
```

```python
#score 550 or less
print(norm.cdf(550,494,100))
```

```python
#between 300 and 600
print(norm.cdf(600,494,100)-norm.cdf(300,494,100))
```

```python
# if area is given and we hv to find value of x in std normal dist
norm.ppf(0.95)
```
