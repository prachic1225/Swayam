```python
data = [1,2,4,23,45,6,45,7,98,90,5,134,54,46756]
def min_and_max(data):
    min_value = min(data)
    max_value = max(data)
    
    return(min_value,max_value)
```

```python
print(min_and_max(data))
```

```python
#Finding Range of the data
```

```python
def range(data):
    min_val=min(data)
    max_val=max(data)

    return(max_val - min_val)
```

```python
range(data)
```
